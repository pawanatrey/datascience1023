```python
# Testo build Linear Regression Model
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/training-ml/Task/main/Test_Sales_LinReg.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
# Fill the Na

data['Republic'] = data['Republic'].fillna(data['Republic'].mean())
data['NDTV'] = data['NDTV'].fillna(data['NDTV'].mean())
data['TV9'] = data['TV9'].fillna(data['TV9'].mean())
data['AajTak'] = data['AajTak'].fillna(data['AajTak'].mean())
```

```python
data.describe()
```

```python
# How data is Distributed

plt.figure(figsize=(20,15), facecolor='yellow')
graphnumber=1

for column in data:
    if graphnumber<=6:
        ax=plt.subplot(2,3,graphnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        
    graphnumber+=1
plt.tight_layout()     
```

```python
# In above graph we observe that Bell Shape is there only with few graph (Repunlic, NDTV). As per the graph we can 
# can observe there is no relation between TV9 & Label and TV5 & Lable so we delete that features 
# So we delete these 
```

```python
# Now We Divide Feature and Lable.

y=data['sales']
x=data.drop(columns=['sales','TV5','TV9'])
```

```python
y
```

```python
x
```

```python
# Now check the relationship using graph or visualize the data (EDA).

plt.figure(figsize=(20,15), facecolor='red')
graphnumber=1

for column in x:
    if graphnumber<=3:
        ax=plt.subplot(1,3,graphnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Sales With Different News Paper',fontsize=20)
        
        
    graphnumber+=1
plt.tight_layout() 
```

```python
# Data Standarization

scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

```python
# Train Test Split

x_train,x_test,y_train,y_test =train_test_split(x_scaled,y,test_size=0.25,random_state=225)
y_train.head()
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
# Now we predict the sale and compare with actual sale,take lat only 

x.tail(4)
```

```python
# Now we predict the sale

Predicted_sale=regression.predict(scaler.transform([[75.5,10.8,75.5]]))
print('Sale by using all Channels investment',Predicted_sale)
```

```python
# Now you can see both predicted and autual are very close
```

```python
# Now we use check how our model is fit with train data using Adjusted R2.

regression.score(x_train,y_train)
```

```python
# Now we use check how our model is fit with test data using Adjusted R2.

regression.score(x_test,y_test)
```

```python
# Now test it by visulized the data

x_test[4] # predict value (18.94931794)
```

```python
y_pred=regression.predict(x_test)
y_pred
```

```python
# Now visualized how prediction and actual look

plt.scatter(y_test,y_pred)
plt.xlabel('Actual Sale')
plt.ylabel('Predicted Sale')
plt.title("Actual vs Model Predicted Sale")
plt.show()
```

```python
# Now Find MAE, MSE, RMSE
from sklearn.metrics import mean_squared_error,mean_absolute_error

mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge,Lasso,LassoCV,RidgeCV
```

```python
# Find the best learning rate

lasscv=LassoCV(max_iter=5) 
lasscv.fit(x_train,y_train)
```

```python
alpha=lasscv.alpha_
alpha
```

```python
# Build the Lasso Model with best Learing Rate

lass_reg=Lasso(alpha)
lass_reg.fit(x_test,y_test)
```
